```python
datatemp = df.filter(['Temperatura'])
dataset = datatemp.values
training_data_len = math.ceil(len(dataset) * .8)
training_data_len
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)
scaled_data
```

```python
#Crea el conjunto de datos de entrenamiento
#Crea el conjunto de escalado de datos
train_data = scaled_data[0:training_data_len, :]
#División de los datos en conjuntos x_train y y_train
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i<=60:
        print(x_train)
        print(y_train)
        print()
```

```python
#Conversión de datos x_train y y_train en matrices Numpy
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
#Reshape data
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
x_train.shape
```

```python
from tensorflow import keras
```

```python
#Build LSTM model
modelo = Sequential()
modelo.add(LSTM(50, return_sequences=True, input_shape= (x_train.shape[1], 1)))
modelo.add(LSTM(50, return_sequences=False))
modelo.add(Dense(25))
modelo.add(Dense(1))
```

```python
#Compilación del modelo
modelo.compile(optimizer='adam', loss='mean_squared_error')
```

```python
#Entreno del model
modelo.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
#Crear el conjunto de datos de prueba
#Cree una nueva matriz que contenga valores escalados del índice
test_data = scaled_data[training_data_len - 60: , :]
#Creación de los data sets x_test y y_test
x_test = []
y_test = dataset[training_data_len:, :]
for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])
```

```python
#Convierto los datos en una matriz numpy
x_test = np.array(x_test)
```

```python
#Reshape de los datos
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1], 1))
```

```python
#Obtención las temperaturas predichas por los modelos
predicciones = modelo.predict(x_test)
predicciones = scaler.inverse_transform(predicciones)
```

```python
#El error cuadrático medio (RMSE) mide la cantidad de error que hay entre dos conjuntos de datos.
rmse = np.sqrt(np.mean(((predicciones - y_test)** 2)))
rmse 
"""Para evaluar las predicciones una de las medidas más utilizadas es la raíz cuadrada
del error cuadrático medio (Root-Mean-Square-Error en inglés [19]). Esta medida es
usada para cuantificar las diferencias entre valores predichos y valores reales. Es una
medida de precisión y se calcula como la raíz cuadrada de la media de los errores al
cuadrado. En forma de ecuación sería:"""
```

```python
rmse
```

```python
#Plot de los datos
train = df[:training_data_len]
valid = df[training_data_len:]
valid['Predicciones'] = predicciones

#Visualización de datos
#plt.plot(datos)
plt.figure(figsize=(14,6))
plt.title('Modelo')
plt.xlabel('Datos', fontsize=16)
plt.ylabel('Temperatura Ibagué', fontsize=16)
plt.plot(train['Temperatura'])
plt.plot(valid[['Temperatura', 'Predicciones']])
plt.legend([['Train', 'Val', 'Predicciones']], loc='lower right')
plt.show()
```
